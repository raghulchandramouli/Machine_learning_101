```python
import pandas as pd
import numpy as np
import warnings
from sklearn import linear_model
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'/kaggle/input/home-prices-dataset/homeprices (2).csv')
df.head(5)
```

```python
median_bedroom = df.bedrooms.median()
median_bedroom
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedroom)
df
```

```python
rg = linear_model.LinearRegression()
# Fit the model
rg.fit(df[['area', 'bedrooms', 'age']], df['price'])
```

```python
rg.coef_
```

```python
rg.intercept_
```

```python
rg.predict([[3000,3,15]])
```
